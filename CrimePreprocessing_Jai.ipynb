```python
import pandas as pd

crime = pd.read_csv(r"\Users\singh\Downloads\Crime_Incidents_4271237485119492148.csv")
```

```python
crime
```

```python
# PREPROCESSING STEP 1
# handle duplicate and null instances
# remove unnecessary rows
# check for any suprising values
```

```python
# Check for duplicate entries

duplicate_rows = crime[crime.duplicated()]
duplicate_rows
```

```python
# Look at rows with null entries

rows_with_null = crime[crime.isnull().any(axis=1)]
rows_with_null
```

```python
# We can see here that there are significantly less LAT and LON values in the rows with null entries. LAT and LON are crucial for our model
# So, remove ONLY the entries with null LAT and LON values first

rows_with_null.info()
```

```python
# Remove the entries with empty LAT and LON values

crime = crime.dropna(subset=['LAT', 'LON'])
```

```python
# Now, LAT and LON are no longer a problem
# Entries with an empty value do not have LAT and LON as empty

rows_with_null = crime[crime.isnull().any(axis=1)]
rows_with_null.info()
```

```python
# Only use entries from years greater than or equal to 2016
# Years before do not have enough values

crime = crime[crime['OffenseYear'] >= 2016]
```

```python
# Save a version of dataframe for temporal and spatial pattern exploration

crime.to_csv("crime_preprocessed_exploration", index=False)
```

```python
# Now, let's think about what attributes we don't need to trian our model
# Model is soley based on LAT and LON, so we can remove all other geographic identifiers
# IMPORTANT: We will remove Census tracts for now, but if we want economic/socioeconomic data we can add an attribute back

cols = ['City', 'Zip', 'Address_Public', 'WARD', 'CENSUS_TRACT', 'CENSUS_TRACT_GEOID', 'CENSUS_BLOCK_GROUP', 'CENSUS_BG_GEOID', 'CENSUS_BLOCK', 'CENSUS_BLOCK_GEOID', 'x', 'y']
crime = crime.drop(columns=cols)
crime.shape[0]
```

```python
# No, empty values anywhere now

rows_with_null = crime[crime.isnull().any(axis=1)]
rows_with_null.info()
```

```python
# We now have a "complete" dataset

crime.info()
```

```python
# Remove the columns we do not need

cols = ['OBJECTID', 'PrimaryKey', 'CaseNumber', 'District', 'ReportedDate', 'DOWname', 'Statute', 'StatDesc']
crime = crime.drop(columns=cols)
crime.info()
```

```python
# Let's take a deeper look into each of the attribute to make sure the data has no gaps or surprises

for col in crime.columns:
    print(crime[col].unique())
    print("\n")
```

```python
# Everything looks good...but let's just make sure that our crucial LAT and LON don't have any non-numeric vales

cols = ['LAT', 'LON']
non_numeric = crime[cols].apply(pd.to_numeric, errors='coerce')
non_numeric_rows = crime[non_numeric.isnull().any(axis=1)]
print(non_numeric_rows)
```

```python
# END OF PREPROCESSING STEP 1
# check for duplicates
# remove null values
# remove unnecessary rows
# make sure there are no suprising values for any attribute
```

```python
# PREPROCESSING STEP 2
# make data ready for model entry
# handle non-numerical data
```

```python
crime.columns
```

```python
# There is are discrepancies between the 'OffenseMonth', 'OffenseDay' and 'OffenseDate'

# Assuming 'OffenseDate' is a datetime string, convert it to a datetime object
crime['OffenseDate'] = pd.to_datetime(crime['OffenseDate'], errors='coerce')

# Extract month and day from the 'OffenseDate' column
crime['ExtractedMonth'] = crime['OffenseDate'].dt.month
crime['ExtractedDay'] = crime['OffenseDate'].dt.day

# Identify rows where OffenseMonth/OffenseDay and OffenseDate don't match
inconsistent_records = crime[(crime['OffenseMonth'] != crime['ExtractedMonth']) | (crime['OffenseDay'] != crime['ExtractedDay'])]
```

```python
# There are 155000 inconsistencies so, we will remove 'OffenseMonth' & 'OffenseDay'
# and use the 'ExtractedMonth' and 'ExtractedDay' attributes instead

inconsistent_records.shape[0]
```

```python
# Drop no longer needed attributes

crime = crime.drop(columns=['OffenseMonth', 'OffenseDay'])
```

```python
crime
```

```python
# Use a label encoder to create dictionaries that map our non-numeric values
# to numeric values

from sklearn.preprocessing import LabelEncoder

mappings = {}

cols = ['UCRdesc', 'OffenseYear', 'TimeGroup', 'TimeBlock']
for col in cols:
    le = LabelEncoder()
    crime[col + '_numeric'] = le.fit_transform(crime[col])
    
    mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))
    
for col, mapping in mappings.items():
    print(f"{col}: {mapping}")
    print("\n")
```

```python
# Drop last remaining unneeded columns
# Only remaining column that may be unnessary is OffenseDate
# It is useful to have a column in datetime format

cols = ['UCRdesc', 'OffenseYear', 'TimeGroup', 'TimeBlock']

crime = crime.drop(columns=cols)
```
