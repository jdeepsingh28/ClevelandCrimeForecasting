```python
import pandas as pd
pd.set_option("display.max_columns", None)
```

```python
crime = pd.read_csv("crime_")
```

```python
violent_crime_list = ['Assault', 'Fel Assault', 'Weapons', 'Robbery', 'Offense Against Family/Children', 'Rape', 'Sex Offenses', 'Homicide']
violent_crime = crime[crime['UCRdesc'].isin(violent_crime_list)]

non_violent_crime_list = ['All Other Offenses', 'Theft', 'Vandalism', 'Burglary', 'GTMV', 'Fraud', 'Drug Abuse Violations', 'Driving Under The Influence', 'Embezzlement', 'Stolen Property', 'Forgery & Counterfeiting', 'Arson', 'Disorderly Conduct', 'Liquor Laws', 'Prostitution', 'Property Damaged', 'Traffic Violations', 'Drunkenness', 'Gambling', '1C']
non_violent_crime = crime[crime['UCRdesc'].isin(non_violent_crime_list)]
```

```python
# ALL CRIME

import matplotlib.pyplot as plt

crime['OffenseDate'] = pd.to_datetime(crime['OffenseDate'])

crime_monthly = crime.resample('M', on='OffenseDate').size()

plt.figure(figsize=(12, 6))
crime_monthly.plot()
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.show()

crime_dow = crime.groupby('DOW').size()

plt.figure(figsize=(8, 5))
crime_dow.plot(kind='bar')
plt.title('Crime by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.show()

crime_hour = crime.groupby('HourofDay').size()

# Plot crime trends by hour of the day
plt.figure(figsize=(8, 5))
crime_hour.plot(kind='bar')
plt.title('Crime by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.show()
```

```python
# VIOLENT CRIME

import matplotlib.pyplot as plt

violent_crime['ReportedDate'] = pd.to_datetime(violent_crime['ReportedDate'])
violent_crime['OffenseDate'] = pd.to_datetime(violent_crime['OffenseDate'])

crime_monthly = violent_crime.resample('M', on='ReportedDate').size()

plt.figure(figsize=(12, 6))
crime_monthly.plot()
plt.title('Monthly Crime Trends')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.show()

crime_dow = violent_crime.groupby('DOW').size()

plt.figure(figsize=(8, 5))
crime_dow.plot(kind='bar')
plt.title('Crime by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Crimes')
plt.show()

crime_hour = violent_crime.groupby('HourofDay').size()

plt.figure(figsize=(8, 5))
crime_hour.plot(kind='bar')
plt.title('Crime by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Crimes')
plt.show()
```
